3. Considera las decisiones de compra de los próximos tres clientes que lleguen a la tienda de ropa Martin Clothing Store. De acuerdo con la experiencia, el gerente de la tienda estima que la probabilidad de que un cliente realice una compra es 0.30. 
  - a. Describa si cumple con las reglas para clasificarlo como un experimiento binomial.
  - b. ¿Cuál es la probabilidad de que dos de los próximos tres clientes realicen una compra?
  - c. ¿Cuál es la probabilidad de que cuatro de los próximos diez clientes realicen una compra?

```python
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt
from math import factorial, e
```

```python
def dist_binomial(n,k,p):
    result = (p**k)*((1-p)**(n-k))*factorial(n)/(factorial(k)*factorial(n-k))
    return result.__round__(4)

print('a) Cumple con las reglas para una distribución binomial: Ensayos idénticos, 2 resultados posibles, probabilidad constante y ensayos independientes.')
print(f'b) La probabilidad es: {dist_binomial(3,2,0.3)}')
print(f'c) La probabilidad es: {dist_binomial(10,4,0.3)}')

# Graficando Binomial
fig, ax = plt.subplots(nrows=1, figsize=(12,5), ncols=2)

N, p = 3, 0.3 # parametros de forma 
binomial = stats.binom(N, p) # Distribución
x = np.arange(start=binomial.ppf(0.01),stop=N+1)
fmp = binomial.pmf(x) # Función de Masa de Probabilidad
ax[0].plot(x, fmp, '--', color='r') #Esta función recibe un conjunto de valores x e y y los muestra en el plano como puntos unidos por línea.
ax[0].set_title('punto B', fontsize='large', fontweight = 600)
ax[0].vlines(x, 0, fmp, colors='r', lw=5, alpha=0.5) #Esta función da formato a las figuras.

N, p = 10, 0.3 # parametros de forma 
binomial = stats.binom(N, p) # Distribución
x = np.arange(start=binomial.ppf(0.01),stop=N+1)
fmp = binomial.pmf(x) # Función de Masa de Probabilidad
ax[1].plot(x, fmp, '--') #Esta función recibe un conjunto de valores x e y y los muestra en el plano como puntos unidos por línea.
ax[1].set_title('punto C', fontsize='large', fontweight = 600)
ax[1].vlines(x, 0, fmp, colors='b', lw=5, alpha=0.5) #Esta función da formato a las figuras.

plt.show() #Esta función muestra las figuras
```

4.  A la oficina de reservaciones de una aerolínea regional llegan 48 llamadas por hora.
- a. Calcule la probabilidad de recibir cinco llamadas en un lapso de 5 minutos.
- b. Estime la probabilidad de recibir exactamente 10 llamadas en un lapso de 15 minutos.
- c. Suponga que no hay ninguna llamada en espera. Si el agente de viajes necesitará 5 minutos para la llamada que está atendiendo, ¿cuántas llamadas habrá en espera para cuando él termine? ¿Cuál es la probabilidad de que no haya ninguna llamada en espera?
- d. Si en este momento no hay ninguna llamada, ¿cuál es la probabilidad de que el agente de viajes pueda tomar 3 minutos de descanso sin ser interrumpido por una llamada?

```python
def dist_poisson(lamba_np,x):
     probabilidad = (pow(e,-lamba_np) * pow(lamba_np,x))/factorial(x)
     return probabilidad.__round__(4)
lamba = 48/60
print(f'a) La probabilidad es: {dist_poisson(lamba*5,5)}')
print(f'b) La probabilidad es: {dist_poisson(lamba*15,10)}')
print('c) Probabilidad de 3 o 4 en espera(misma probabilidad):',dist_poisson(lamba*5,4),'\n','  Probabilidad de 0 en espera:', dist_poisson(lamba*5,0))
print(f'd) La probabilidad es: {dist_poisson(lamba*3,0)}')

# Graficando Poisson

fig, ax = plt.subplots(nrows=1, figsize=(18,5), ncols=3)
mu =  lamba*5 # parametro de forma 
poisson = stats.poisson(mu) # Distribución
x = np.arange(poisson.ppf(0.01),
              poisson.ppf(0.99))
fmp = poisson.pmf(x) # Función de Masa de Probabilidad
ax[0].plot(x, fmp, '--')
ax[0].vlines(x, 0, fmp, colors='black', lw=5, alpha=0.5)
ax[0].set_title('punto A y C', fontsize='large', fontweight = 600)
ax[0].grid()

mu =  lamba*15 # parametro de forma 
poisson = stats.poisson(mu) # Distribución
x = np.arange(poisson.ppf(0.01),
              poisson.ppf(0.99))
fmp = poisson.pmf(x) # Función de Masa de Probabilidad
ax[1].plot(x, fmp, '--')
ax[1].vlines(x, 0, fmp, colors='b', lw=5, alpha=0.5)
ax[1].set_title('punto B', fontsize='large', fontweight = 600)
ax[1].grid()

mu =  lamba*3 # parametro de forma 
poisson = stats.poisson(mu) # Distribución
x = np.arange(poisson.ppf(0.01),
              poisson.ppf(0.99))
fmp = poisson.pmf(x) # Función de Masa de Probabilidad
ax[2].plot(x, fmp, '--')
ax[2].vlines(x, 0, fmp, colors='r', lw=5, alpha=0.5)
ax[2].set_title('punto D', fontsize='large', fontweight = 600)
ax[2].grid()

plt.show()
```

5. En una encuesta realizada por Gallup Organization, se les preguntó a los interrogados, “Cuál es el deporte que prefieres ver”. Futbol y básquetbol ocuparon el primero y segundo lugar de preferencia (www.gallup.com, 3 de enero de 2004). Si en un grupo de 10 individuos, siete prefieren futbol y tres prefieren básquetbol. Se toma una muestra aleatoria de tres de estas personas.
- a. ¿Cuál es la probabilidad de que exactamente dos prefieren el futbol?
- b. ¿De que la mayoría (ya sean dos o tres) prefiere el futbol?

```python
def dist_hipergeometrica(N,X,n,x):
  Xx = factorial(X)/(factorial(x)*factorial(X-x))
  NX_nx= factorial(N-X)/(factorial(n-x)*factorial((N-X)-(n-x)))
  Nn = factorial(N)/(factorial(n)*factorial(N-n))
  hipergeometrica = (Xx * NX_nx)/Nn

  return hipergeometrica.__round__(4)

print(f'a) La probabilidad es: {dist_hipergeometrica(10,7,3,2)}')
print(f'b) Es la suma de que 2 prefieran ({dist_hipergeometrica(10,7,3,2)}) y 3 lo prefieran ({dist_hipergeometrica(10,7,3,3)}). La probabilidad es: {dist_hipergeometrica(10,7,3,2)+dist_hipergeometrica(10,7,3,3)}')
M, n, N = 10, 3, 7 # parametros de forma 
hipergeometrica = stats.hypergeom(M, n, N) # Distribución
x = np.arange(0, n+1)
fmp = hipergeometrica.pmf(x) # Función de Masa de Probabilidad
plt.plot(x, fmp, '--')
plt.vlines(x, 0, fmp, colors='b', lw=5, alpha=0.5)
plt.title('Distribución Hipergeométrica')
plt.grid()
plt.ylabel('Probabilidad')
plt.xlabel('Casos')
plt.show()
```

13. La calificación promedio de los estudiantes de Henry es de 78 con una desviación estandar de 25. Cual es la probabilidad de tener estudiantes con calificaciones mayores o iguales 90.

```python
Z = (90 - 78)/25
print(f'Para el valor de {Z=}, la probabilidad es de {1-0.6844}')
# Graficando Normal
mu, sigma = 78, 25 # media y desvio estandar
normal = stats.norm(mu, sigma)
x = np.linspace(normal.ppf(0.01),
                normal.ppf(0.99), 100)
fp = normal.pdf(x) # Función de Probabilidad
plt.plot(x, fp)
plt.title('Distribución Normal')
plt.grid()
plt.ylabel('Probabilidad')
plt.xlabel('Casos')
plt.show()
```
